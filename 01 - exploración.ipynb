### <font color='46B8A9'> **2.2. Distribución y participación de variables categóricas**

Número de estudiantes por clase en cada variable categórica:

```python
'''Para cada variable categórica se crea un DataFrame con los valores únicos de la columna como índice, donde la primera
   columna corresponde a la frecuencia absoluta (número de veces que se repite cada valor o categoría), y la segunda columna
   muestra la frecuencia relativa (porcentaje que representa cada categoría respecto al total de observaciones en esa columna)'''

columnas1 = df.select_dtypes(include=['object'])  # Selecciona las columnas categóricas (tipo 'object') del DataFrame

# Crear un nuevo DataFrame con frecuencia absoluta y frecuencia relativa de cada categoría
for columna in columnas1:
    df_count = pd.DataFrame({ "Conteo": df[columna].value_counts(),
                             "Proporción": df[columna].value_counts(normalize=True).mul(100).round(2).astype(str) + ' %'})
    print(df_count)  # Imprime el DataFrame con las frecuencias
    print('-'*70)
```

### <font color='46B8A9'> **2.6. Análisis relevantes**

Debido a que se quiere predecir el desempeño que va a tener cada estudiante, se plantea que la variable 'ESTU_HORASSEMANATRABAJA', que indica las horas a la semana que trabaja, puede incidir en el desempeño del estudiante en la prueba. En las próximas entregas se espera analizar más detalladamente el comportamiento de las variables que componen el DataFrame.

```python
import plotly.express as px

# Contar los valores únicos en la columna 'ESTU_HORASSEMANATRABAJA' y convertir el resultado en un DataFrame
df_count = df['ESTU_HORASSEMANATRABAJA'].value_counts().reset_index()

# Renombrar las columnas del DataFrame resultante
df_count.columns = ['ESTU_HORASSEMANATRABAJA', 'count']

# Ordenar el DataFrame de mayor a menor según la cantidad de estudiantes
df_count = df_count.sort_values('count', ascending=False)

# Crear gráfico de barras
fig = px.bar(
    df_count,
    x='ESTU_HORASSEMANATRABAJA',
    y='count',
    text='count',
    labels={
        'ESTU_HORASSEMANATRABAJA': 'Horas que trabaja a la semana',
        'count': 'Cantidad de estudiantes'
    },
    title='<b>Distribución de estudiantes según las horas que trabajan a la semana</b>'
)

# Configurar diseño del gráfico
fig.update_traces(textposition='outside')
fig.update_layout(
    xaxis_title='Horas que trabaja a la semana',
    yaxis_title='Número de estudiantes',
    title_x=0.5,
    coloraxis_showscale=False
)

# Mostrar el gráfico
fig.show()
```
